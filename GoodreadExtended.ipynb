```
# Reviews for the books from GoodReads
import requests
from bs4 import BeautifulSoup, Comment
import pandas as pd
import re
import matplotlib.pyplot as plt

url='https://www.goodreads.com/search?utf8=%E2%9C%93&query=Robin+Sharma'
```

```
ins_company = []

#Loop through our URLs we loaded above
html = requests.get(url).text
soup = BeautifulSoup(html, "html.parser")
#identify table we want to scrape
ins_table = soup.find('table')

#try clause to skip any  books with missing/empty board data tables
try:
#loop through table, grab each of the 4 columns shown (try one of the links yourself to see the layout)
    for row in ins_table.find_all('tr'):
        cols = row.find_all('a')
        #if len(cols) == 4:
        #a = cols.find_all('a',attrs={'itemprop':'name'})
        rate1 = row.find_all('span',attrs={"class":"minirating"})
        ins_company.append((cols[1].text.strip(), cols[2].text.strip(),cols[3].text.strip(),rate1[0].text.strip()))
except: pass  
```

```
import numpy as np
ins_array=np.asarray(ins_company)
len(ins_array)
ins_array
```

```
#convert new array to dataframe
df = pd.DataFrame(ins_array,columns=['Books','Author','Edition','Rating for the Book'])
df
#df.reindex(df.index.drop(0))
```
